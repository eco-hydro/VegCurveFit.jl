```julia
n = 8
D = ddmat(1:n, 3)
display(D)

M = diagm(ones(n)) + λ * D' * D
```

```julia
U = deepcopy(M)
L = typeof(U)(diagm(ones(n)))

# 仅运行三次，猜测公式的形式
for i = 1:3
    r1 = U[i, :]
    for j = i+1:min(i+2, n)
        f = U[j, i] / U[i, i]
        L[j, i] = f
        U[j, :] .= U[j, :] .- (f * r1)
    end
end
## U
## Elier采用的是LU分解 
display(L)
display(U)
```
